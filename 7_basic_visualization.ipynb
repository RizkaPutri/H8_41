```python
df_canada = pd.read_excel("https://github.com/ardhiraka/PFDS_sources/blob/master/Canada.xlsx?raw=true",
    sheet_name="Canada by Citizenship",
    skiprows = range(20),
    skipfooter = 2
)
```

```python
df_canada.sort_values(by = 'Total', ascending = False, axis = 0, inplace=True)
```

```python
top5_countries = df_canada.sort_values(
    by = "Total", ascending = False, axis = 0
).head()
top5_countries = top5_countries.set_index("Country")
top5_countries = top5_countries[years].transpose()
```

```python
top5_countries.head()
```

```python
top5_countries.plot(figsize=(15,9))
```

```python
top5_countries.plot(kind = 'area',
                   stacked = False,
                   figsize=(10,10))

plt.ylabel("Number of Imigration from Top 5 Country")
plt.xlabel("Years")
plt.title("Imigration from Top 5 Country over Years")
```

```python
top5_countries.plot(kind = 'area',
                    alpha = 0.25,
                   stacked = False,
                   figsize=(10,10))

plt.ylabel("Number of Imigration from Top 5 Country")
plt.xlabel("Years")
plt.title("Imigration from Top 5 Country over Years")
```

```python
ax = top5_countries.plot(kind = 'area',
                    alpha = 0.25,
                   stacked = False,
                   figsize=(10,10))

ax.set_ylabel("Number of Imigration from Top 5 Country")
ax.set_xlabel("Years")
ax.set_title("Imigration from Top 5 Country over Years")
```
